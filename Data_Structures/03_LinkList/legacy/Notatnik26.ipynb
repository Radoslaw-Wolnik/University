```python
# Zadanie 1

class Node():
    def __init__(self, item = None):
        self.item = item
        self.next = None
        
        
class LinkList():
    
    def __init__(self):
        self.head = None
        self.size = 0
        
    def isEmpty(self):
        return self.size == 0
    
    def addLast(self, item):
        new = Node(item)
        if self.isEmpty():
            self.head = new
        else:
            current = self.head
            while current.next != None:
                current = current.next
            current.next = new
        self.size +=1

    # nie dziala dlatego ze usuwa referencje do miejsca w pamieci a nie wezel
    # def delLast(self):
    #     lastNode = self.head
    #     while lastNode is not None:
    #         lastNode = lastNode.next
    #     del lastNode
        
    def delLast(self):
        lastNode = self.head
        prevNode = None
        while lastNode.next is not None:
            prevNode = lastNode
            lastNode = lastNode.next
        prevNode.next = None
        self.size -= 1
        return lastNode
        
    def addFirst(self, item):
        new = Node(item)
        x = self.head
        new.next = x
        self.head = new
        self.size +=1
        
    def removeFirst(self):
        res = self.head.item
        x = self.head.next
        self.head = x
        self.size -= 1
        return res
    
    
    def insertAt(self, val, position):
        if position > self.size or position == -1:
            self.addLast(val)
        elif position == 0:
            self.addFirst(val)
        else:
            NewNode = Node(val)
            iterator = 0
            currentNode = self.head
            while True:
                if iterator == position:
                    NewNode.next = currentNode
                    prevnode.next = NewNode
                    break
                prevnode = currentNode
                currentNode = currentNode.next
                iterator +=1
        self.size += 1
    
    def delAt(self, position):
        if position > self.size or position == -1:
            self.size -= 1
            return self.delLast()
        elif position == 0:
            self.size -= 1
            return self.removeFirst()
        else:
            res = None
            iterator = 0
            prevnode = None
            currentNode = self.head
            while True:
                if iterator == position:
                    res = currentNode
                    prevnode.next = currentNode.next
                    break
                prevnode = currentNode
                currentNode = currentNode.next
                iterator +=1
            self.size -= 1
            return res
    
    
    def __str__(self):
        print_val = self.head
        ret = []
        while print_val != None:
            ret.append(print_val.item)
            print_val = print_val.next
        return str(ret)
    
```

```python
# zadanie 5
# Stos in linked list

class StackLink():
    
    def __init__(self):
        self.core = LinkList()
    
    def __str__(self):
        return str(self.core)
    
    def pop(self):
        return self.core.removeFirst()
        
    def push(self, item):
        self.core.addFirst(item)
        
    def isEmpty(self):
        return self.core.size == 0
```

```python
test = StackLink()
example = [1,2,3,4,5,6,7,'a']

for el in example:
    test.push(el)
    print(test)
print()

while test.isEmpty() == False:
    x = test.pop()
    print(f'pop = {x}, test = {test}')
print('\n')
print(test)
print()

for el in example:
    test.push(el)
print(test)
for _ in range(3):
    test.pop()
print(test)
```

```python
# zadanie 7
# kolejka Fifo in lista cykliczna
class Queue():
    
    def __init__(self, n = 10):
        self.items = n* [None]
        self.size = 0
        self._size = n
        self.head = 0
        self.tail = 0
        
    def isEmpty(self):
        return self.size == 0
    
    def isFull(self):
        return self._size == self.size
    
    def enqueue(self, item):
        if self.isFull():
            raise ValueError('Fifo is Full')
        self.items[self.tail] = item
        self.size += 1
        if self.tail == self._size - 1:
            self.tail = 0
        else:
            self.tail += 1
            
    def dequeue(self):
        if self.isEmpty():
            raise ValueError('Fifo is Empty')
        res = self.items[self.head]
        self.items[self.head] = None
        self.size -= 1
        if self.head == self._size - 1:
            self.head = 0
        else:
            self.head += 1
        return res
    
    def __str__(self):
        return str(self.items)
```

```python
Try01 = Queue(10)
example = [1,2,3,4,5,6,7,'a']

for el in example:
    Try01.enqueue(el)
print(Try01)
print()

while Try01.isEmpty() == False:
    print(Try01.dequeue(), ' ', end = '') # end = ''
print('\n')
print(Try01)
print()

for el in example:
    Try01.enqueue(el)
print(Try01)
for _ in range(3):
    Try01.dequeue()
print(Try01)

print('\n'*2)

Try02 = Queue(2)
example = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
for el in example:
    print(Try02)
    Try02.enqueue(el)
```

```python
# zadanie 8
class Stack:
    def __init__(self):
        self.start()
    
    def start(self):
        self.structure = []
            
    def push(self, val):
        assert isinstance(val, (str, int, float)), "This element can't be in Stack\nU can add str, int or float"
        self.structure.append(val)
    
    def pop(self):
        if self.is_empty():
            return 'Stack is empty'
        return self.structure.pop()
    
    def peak(self):
        if self.is_empty():
            return 'Stack is empty'
        return self.structure[-1]
            
    def __str__(self):
        return str(self.structure)
    
    def is_empty(self):
        if len(self.structure) == 0:
            return True
        else:
            return False
        
        
# zadanie 1 zestaw 3
class Kolejka_2_stosy():
    def __init__(self):
        self.One = Stack()
        self.Two = Stack()
    
    def enqueue(self, val):
        self.One.push(val)
        
    def dequeue(self):
        res = None
        while self.One.peak() != 'Stack is empty':
            self.Two.push(self.One.pop())
        res = self.Two.pop()
        while self.Two.peak() != 'Stack is empty':
            self.One.push(self.Two.pop())
        return res
    
Kolejka = Kolejka_2_stosy()
Kolejka.enqueue('E')
Kolejka.enqueue(1)
Kolejka.enqueue(2)
Kolejka.enqueue(3)
print(Kolejka.dequeue())
print(Kolejka.dequeue())
print(Kolejka.dequeue())
```

```python
# zadanie 9
# a
# Fifo dwustronne in tablica cykliczna
# [][][][head][1][2][3][4][tail][][][]

class FiLiFoLo():
    
    def __init__(self, n = 10):
        self.items = n* [None]
        self.size = 0
        self._size = n
        self.head = 0
        self.tail = 0
        
    def isEmpty(self):
        return self.size == 0
    
    def isFull(self):
        return self._size == self.size
    
    def enqueueLast(self, item):
        if self.isFull():
            raise ValueError('Fifo is Full')
        self.items[self.tail] = item
        self.size += 1
        if self.size == 1: # tablica byla wczesniej pusta
            if self.tail == 0:
                self.head = self._size -1
            else:
                self.head = self.tail - 1
        if self.tail == self._size - 1:
            self.tail = 0
        else:
            self.tail += 1
            
    def enqueueFirst(self, item):
        if self.isFull():
            raise ValueError('Fifo is Full')
        self.items[self.head] = item
        self.size += 1
        if self.size == 1: # tablica byla wczesniej pusta
            if self.head == self._size - 1:
                self.tail = 0
            else:
                self.tail = self.head + 1
        if self.head == 0:
            self.head = self._size -1
        else:
            self.head -= 1
            
    def dequeueFirst(self):
        if self.isEmpty():
            raise ValueError('Fifo is Empty')
        if self.head == self._size -1:
            self.head = 0
        else:
            self.head += 1
        res = self.items[self.head]
        self.items[self.head] = None
        self.size -= 1
        return res
    
    def dequeueLast(self):
        if self.isEmpty():
            raise ValueError('Fifo is Empty')
        if self.tail == 0:
            self.tail = self._size - 1
        else:
            self.tail -= 1
        res = self.items[self.tail]
        self.items[self.tail] = None
        self.size -= 1
        if self.tail == 0:
            self.tail = self._size -1
        else:
            self.tail -= 1
        return res
    
    def __str__(self):
        return str(self.items)
```
