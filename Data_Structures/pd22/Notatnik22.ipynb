```python
# zadanie 1

def wstawienie_sort(L):
    for marker in range(1,len(L)):
        for i in range (marker-1,-1,-1):
            if L[i] > L[marker]:
                L[i], L[marker] = L[marker], L[i]
                marker -= 1
            else:
                break
    return L



def wstawienie_sort_while(L):
    j = 1
    while j < len(L):
        marker = j
        i = marker-1
        while i > -1:
            if L[i] > L[marker]:
                L[i], L[marker] = L[marker], L[i]
                marker -= 1
            else:
                break
            i -= 1
        j += 1
    return L
                
    
print(wstawienie_sort([5, 6, -1, 0, 4, 7, 2, 3, 102, 88]))
print(wstawienie_sort([i for i in range (10)]))
print(wstawienie_sort([i for i in range (10,0,-1)]))        
        
print(wstawienie_sort_while([5, 6, -1, 0, 4, 7, 2, 3, 102, 88]))
print(wstawienie_sort_while([i for i in range (10)]))
print(wstawienie_sort_while([i for i in range (10,0,-1)]))
```

```python
# zadanie 2

import time
import matplotlib.pyplot as plt
from random import choice


def create_plt(sorting_algorytm):
    '''jako algorytm funkcji przekazywana jest odpowiednia funkcja sortujaca'''
    N = [10, 100, 300, 500, 700]
    #N =[2,3,4,10]
    
    tries = 10
    optymistic_plt = []
    pesymistic_plt = []
    average_plt = []
        
    for lenght in N:
        
        optymistyczny = [el for el in range(lenght)]
        pesymistyczny = [lenght - el - 1 for el in range(lenght)]
        sredni = [choice(optymistyczny) for _ in range(lenght)]
        
        optymistic_avr = 0
        pesymistic_avr = 0
        average_avr = 0
        for _ in range(tries):
            start = time.time()
            sorting_algorytm(optymistyczny)
            end = time.time()
            optymistic_avr += end - start
            
            start = time.time()
            sorting_algorytm(pesymistyczny)
            end = time.time()
            pesymistic_avr += end - start
            
            start = time.time()
            sorting_algorytm(sredni)
            end = time.time()
            average_avr += end - start
            
        optymistic_plt.append(optymistic_avr/tries)
        pesymistic_plt.append(pesymistic_avr/tries)
        average_plt.append(average_avr/tries)
        
    plt.figure(figsize=(8, 5))
    plt.plot(N, optymistic_plt, '-', color  = 'skyblue', label = 'optymistyczna sytuacja')
    plt.plot(N, pesymistic_plt, '-', color  = 'crimson', label = 'pesymistyczna sytuacja')
    plt.plot(N, average_plt, '-', color  = 'green', label = 'srednia sytuacja')
    plt.legend(bbox_to_anchor=(1.04,0.5), loc='center left', shadow=True, fontsize='x-large')
    plt.xlabel('dlugosc listy')
    plt.ylabel('czas [s]')
    plt.title(sorting_algorytm.__name__)
    plt.show()
        
create_plt(wstawienie_sort)
create_plt(wstawienie_sort_while)
```

```python
# zadanie 3

def nierosnace_wstawienie_sort(L):
    for marker in range(1,len(L)):
        for i in range (marker-1,-1,-1):
            if L[i] < L[marker]:
                L[i], L[marker] = L[marker], L[i]
                marker -= 1
            else:
                break
    return L



def nierosnace_wstawienie_sort_while(L):
    j = 1
    while j < len(L):
        i = j
        while i != 0 and L[i-1] < L[i]:
            L[i], L[i-1] = L[i-1], L[i]
            i-=1
        j += 1
    return L
                
    
print(nierosnace_wstawienie_sort([5, 6, -1, 0, 4, 7, 2, 3, 102, 88]))
print(nierosnace_wstawienie_sort([i for i in range (10)]))
print(nierosnace_wstawienie_sort([i for i in range (10,0,-1)]))        
print()
print(nierosnace_wstawienie_sort_while([5, 6, -1, 0, 4, 7, 2, 3, 102, 88]))
print(nierosnace_wstawienie_sort_while([i for i in range (10)]))
print(nierosnace_wstawienie_sort_while([i for i in range (10,0,-1)]))
```

```python
# zadanie 4

def insertion_sort(L, kolejnosc):
    if kolejnosc == 'up':
        for i in range(1,len(L)):
            while L[i-1] > L[i] and i > 0:
                L.insert(i-1, L.pop(i))
                i-=1
                
    elif kolejnosc == 'down':
        for i in range(1,len(L)):
            while i != 0 and L[i-1] < L[i]:
                L[i], L[i-1] = L[i-1], L[i]
                i-=1
                
        
    else:
        return ':o'
    return L

print('up')
print(insertion_sort([5, 6, -1, 0, 4, 7, 2, 3, 102, 88], 'up'))
print(insertion_sort([i for i in range (10)], 'up'))
print(insertion_sort([i for i in range (10,0,-1)], 'up'))

print()

print('down')
print(insertion_sort([5, 6, -1, 0, 4, 7, 2, 3, 102, 88], 'down'))
print(insertion_sort([i for i in range (10)], 'down'))
print(insertion_sort([i for i in range (10,0,-1)], 'down'))
```

```python
# zadanie 4 v2

def insertion_sort_v2(Lista, loop, order):
    '''loop = for or while
       order = ascending or descend'''
    if loop == 'for':
        if order == 'ascend':
            return wstawienie_sort(Lista)
        else:
            return nierosnace_wstawienie_sort(Lista)
    else: # while
        if order == 'ascend':
            return wstawienie_sort_while(Lista)
        else:
            return nierosnace_wstawienie_sort_while(Lista)

List01 = [21, 24, 41, 42, 96, 68, 89, 82, 29, 45, 11, 3, 24, 29, 23, 47, 85, 49, 25, 79, 59]
print(insertion_sort_v2(List01, 'for', 'ascend'))
print(insertion_sort_v2(List01, 'for', 'descend'))
print(insertion_sort_v2(List01, 'while', 'ascend'))
print(insertion_sort_v2(List01, 'while', 'descend'))
```

```python
# zadanie 5

def bubble_sort(L):

    sortowanie = True
    licznik = 1
    while sortowanie:
        sortowanie = False
        for i in range(len(L)-licznik):
            if L[i+1] < L[i]:
                L[i+1], L[i] = L[i], L[i+1]
                sortowanie = True
        licznik += 1

    return L
        

create_plt(bubble_sort)
print(bubble_sort([5, 6, -1, 0, 4, 7, 2, 3, 102, 88]))
print(bubble_sort([i for i in range (10)]))
print(bubble_sort([i for i in range (10,0,-1)]))
```

```python
# zadanie 6
# search smallest than take it to the beggining

# optymistic O(n^2)
# average O(n^2)
# pesymistic O(n^2)

def selection_sort(L):
    for i in range(len(L)):
        min = i
        for j in range(i+1,len(L)):
            if L[j] < L[min]:
                min = j
        L[i], L[min] = L[min], L[i]
        return L
        
create_plt(selection_sort)
print(selection_sort([5, 6, -1, 0, 4, 7, 2, 3, 102, 88]))
print(selection_sort([i for i in range (10)]))
print(selection_sort([i for i in range (10,0,-1)]))
```

```python
# zadanie 7
# wyszukiwanie liniowe

def linear_serach(L, search):
    for i in range(len(L)):
        if L[i] == search:
            return i
    return None


# wyszukiwanie liniowe z wartowenikiem 
# (jesli nie znajdzie liczby wyszukiwanej to zwraca wartownik)
# -1

def wartownik_linear_search(L, search):
    L.append(-1)
    for i in range(len(L)):
        if L[i] == search:
            break
            
    if i == len(L)-1:
        return L[i]
    return i


print(linear_serach([2,3,4,7,81,3,8,5,232,393,0], 393))
print(wartownik_linear_search([2,3,4,7,81,3,8,5,232,393,0], 393))

print(linear_serach([2,3,4,7,81,3,8,5,232,393,0], 10000))
print(wartownik_linear_search([2,3,4,7,81,3,8,5,232,393,0], 10000))

print(linear_serach([2,3,4,7,81,3,8,5,232,393,0], 2))
print(wartownik_linear_search([2,3,4,7,81,3,8,5,232,393,0], 2))
```

```python
# zadanie 8
# merge sort

# pesymistycznie O(nlogn)
#
# optymistycznie O(nlogn)

def merge(left, right):
    result = []
    #print(left)
    #print(right)
    while len(left) > 0 and len(right) > 0:
        if left[0] > right[0]:
            result.append(right[0])
            del right[0]
        else:
            result.append(left[0])
            del left[0]
   
    result += left + right 
    #print(result)
    return result


def merge_sort(L):
    '''funkcja sortuje liste za pomoca algorytmu scalenia
    przyjmuje liste'''
    
    
    if len(L) > 1:
        mid = len(L)//2
        left = L[:mid]
        right = L[mid:]
        left = merge_sort(left)
        right = merge_sort(right)
        return merge(left, right)
    else:
        return L

print(merge_sort([3,5,-1,2,7,4]))
print(merge_sort([-3,4,5,7,2,4,-8,0,-3]))
print(merge_sort([5, 6, -1, 0, 4, 7, 2, 3, 102, 88]))
print(merge_sort([i for i in range (10)]))
print(merge_sort([i for i in range (10,0,-1)]))
```

```python
# zadanie 8
# quick sort recursive
from random import randint
# pesymistycznie n^2 chyba
# średnio nlogn chyba
# optymistycznie n

def quicksort_r(L, pivotPoint = 'a'):
    '''funkcja sortuje tablicę rekurencyjnie za pomocą algorytmu quicksort
    L = lista do posortowania
    pivotPoint = (a) początku, (b) końcu, (c) w środku lub w (d) losowym miejscu listy'''

    less = []
    equal = []
    greater = []

    if len(L) > 1:
        pivotdict = {'a' : 0, 'b' : len(L)-1, 'c' : len(L)//2, 'd' : randint(0,len(L)-1)}
        pivot = L[pivotdict[pivotPoint]]
#       print(pivot)
        for el in L:
            if el < pivot:
                less.append(el)
            elif el == pivot:
                equal.append(el)
            elif el > pivot:
                greater.append(el)
        return quicksort_r(less, pivotPoint) + equal + quicksort_r(greater, pivotPoint)
    
    else:
        return L
    

print(quicksort_r([5, 6, -1, 0, 4, 7, 2, 3, 102, 88], 'a'))
print(quicksort_r([5, 6, -1, 0, 4, 7, 2, 3, 102, 88], 'b'))
print(quicksort_r([randint(1,100) for i in range (10)], 'c'))
print(quicksort_r([randint(1,100) for i in range (20)], 'd'))
```

```python
# zadanie 8
# quiick sort iterative

from random import randint
from copy import deepcopy
# pesymistycznie O(n^2)
# średnio O(nlogn)
# optymistycznie O(nlogn)

def quicksort_i(L, pivotPoint = 'a'):
    '''funkcja sortuje tablicę iteracyjnie za pomocą algorytmu quicksort
    L = lista do posortowania
    pivotPoint = (a) początku, (b) końcu, (c) w środku lub w (d) losowym miejscu listy'''
    
    warunek = True
    lista = [L]
    
    while warunek:
#        print(lista)
        warunek = False
        for i in range (len(lista)):
            if len(lista[i]) > 1:
                warunek = True
                break
        
        if len(lista[i]) > 1:
            pivotdict = {'a' : 0, 'b' : len(lista[i])-1, 'c' : len(lista[i])//2, 'd' : randint(0,len(lista[i])-1)}        
            pivot = lista[i][pivotdict[pivotPoint]]

            lower = []
            equal = []
            higher = []

            for el in lista[i]:
                if el < pivot:
                    lower.append(el)
                elif el == pivot:
                    equal.append(el)
                else:
                    higher.append(el)

            del lista[i]
            if len(higher) > 0:
                lista.insert(i, higher)
            if len(equal) > 1:
                for el in equal:
                    lista.insert(i, [el])
            else:
                lista.insert(i, equal)
            if len(lower) > 0:
                lista.insert(i,lower)
            
    result = []
    for el in lista:
        result.append(el[0])
    return result


try01 = [21, 24, 41, 42, 96, 68, 89, 82, 29, 45, 11, 3, 24, 29, 23, 47, 85, 49, 25, 79, 59, 13, 20, 86, 88, 72, 47, 18, 99, 42]
print(quicksort_i([5, 6, -1, 0, 4, 7, 2, 3, 102, 88]))
print(quicksort_i([5, 6, -1, 0, 4, 7, 2, 3], 'a'))
print(quicksort_i(try01, 'a'))

print(quicksort_i([5, 6, -1, 0, 4, 7, 2, 3, 102, 88], 'a'))
print(quicksort_i([5, 6, -1, 0, 4, 7, 2, 3, 102, 88], 'b'))
print(quicksort_i([randint(1,100) for i in range (10)], 'c'))
print(quicksort_i([randint(1,100) for i in range (20)], 'd'))
```
