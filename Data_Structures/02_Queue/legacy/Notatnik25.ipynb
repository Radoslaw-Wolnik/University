```python
class Stack:
    def __init__(self):
        self.start()
    
    def start(self):
        self.structure = []
            
    def add(self, val):
        assert isinstance(val, (str, int, float)), "This element can't be in Stack\nU can add str, int or float"
        self.structure.append(val)
    
    def get(self):
        if self.is_empty():
            return 'Stack is empty'
        return self.structure.pop()
    
    def peak(self):
        if self.is_empty():
            return 'Stack is empty'
        return self.structure[-1]
            
    def __str__(self):
        return str(self.structure)
    
    def is_empty(self):
        if len(self.structure) == 0:
            return True
        else:
            return False
```

```python
# zadanie 1
# "20 10 + 75 45 - *"
# 
# 
def count_postfix(exp):
    operators = ['+', '-', '*', '/']
    Res = Stack()
    exp = exp.split()
    
    for el in exp:
        if el in operators and Res.is_empty() == False:
            b = Res.get()
            a = Res.get()
            if el == '+':
                a = a + b
            elif el == '-':
                a = a - b
            elif el == '/':
                a = a/b
            elif el == '*':
                a = a*b
            Res.add(a)
        
        else:
            Res.add(float(el))
            
    return Res.get()
    
    
x = "20 10 + 75 45 - *"
count_postfix(x)
```

```python
# zadanie 2
def infix_to_postfix(exp):
    operators = ['+', '-', '*', '/']
    res = ''
    Stos = Stack()
    exp = exp.split(' ')
    for element in exp:
        if element == '(':
            pass
            # nothing?
            
        elif element == ')':
            if Stos.is_empty() == False:
                res += Stos.get() + ' '
                
        elif element in operators:
            Stos.add(element)
            
        else:
            res += element + ' '
            
    return res

x = '( 10 + 2 )'
y = '( ( 10 + 2 ) * 3 )'
z = '( ( 10 + 2 ) * ( 2 + 3 ) )'
print(infix_to_postfix(x))
print(infix_to_postfix(y))
print(infix_to_postfix(z))
```

```python
# zadanie 3
def PostfixtoInfix(exp):
    operators = ['+', '-', '*', '/']
    Stos = Stack()
    exp = exp.split(' ')[:-1]
    for element in exp:
        if element in operators:
            b = Stos.get()
            a = Stos.get()
            Stos.add('(' + a + element + b + ')')
        else:
            Stos.add(element)
    return Stos.get()
            
x = '10 2 + '
y = '10 2 + 3 * '
z = '10 2 + 2 3 + * '
print(PostfixtoInfix(x))
print(PostfixtoInfix(y))
print(PostfixtoInfix(z))
```

```python
# zadanie 5
# Fifo in tablica cykliczna
from copy import deepcopy

class Kolejka:
    # self.data structure
    # self.len lenght of structure
    # self.size number of elements in structure

    def __init__(self, lenght = 10):
        self.set_len(lenght) # 10 = default size
        self.MakeNull()
        self.front = 0
        
    def set_len(self, val):
        assert isinstance(val, int), 'Lenght of structure have to be an intiger value'
        self.len = val
        
    def MakeNull(self):
        self.data = [None] *self.len
        self.size = 0
        
    def Front(self):
        return self.data[self.front]
    
    def Empty(self):
        if sum([1 for el in self.data if el != None]) == 0:
            return True
        return False
    
    def Enqueue(self, val): # adds element to structure; add at the end
        position = (self.front + self.size) % self.len
        if self.data[position] != None:
            self.Resize()
            position = (self.front + self.size) % self.len
        self.data[position] = val
        self.size += 1
        
    def Dequeue(self): # returns element from structure; returns the element that was on structure longest time
        val = self.data[self.front]
        self.data[self.front] = None
        self.front = (self.front + 1) % self.len
        self.size -= 1
        return val
    
    def Resize(self):
        temp = deepcopy(self)
        self.set_len(self.len *2)
        self.MakeNull()
        while temp.Empty() == False:
            self.Enqueue(temp.Dequeue())
            
    def Resize2(self):
        temp = self.data
        walk = self.Front()
        self.set_len(self.len *2)
        self.MakeNull()
        for i in range(len(temp)):
            self.Enqueue(temp[walk])
            walk = (walk+1) % len(temp)
        
        
    def __str__(self):
        return f'front: {self.front} size: {self.size} lenght: {self.len}\n' + f'{self.data}'
```

```python
Try01 = Kolejka()
example = [1,2,3,4,5,6,7,'a']

for el in example:
    Try01.Enqueue(el)
print(Try01)
print()

while Try01.Empty() == False:
    print(Try01.Dequeue(), ' ', end = '') # end = ''
print('\n')
print(Try01)
print()

for el in example:
    Try01.Enqueue(el)
print(Try01)
for _ in range(3):
    Try01.Dequeue()
print(Try01)
```

```python
# zadanie 6
# Link Lists

class Node():
    def __init__(self, item = None):
        self.item = item
        self.next = None
        
        
class LinkList():
    
    def __init__(self):
        self.head = None
        self.size = 0
        
    def isEmpty(self):
        return self.size == 0
    
    def addLast(self, item):
        new = Node(item)
        if self.isEmpty():
            self.head = new
        else:
            current = self.head
            while current.next != None:
                current = current.next
            current.next = new
        self.size +=1

    # nie dziala dlatego ze usuwa referencje do miejsca w pamieci a nie wezel
    # def delLast(self):
    #     lastNode = self.head
    #     while lastNode is not None:
    #         lastNode = lastNode.next
    #     del lastNode
        
    def delLast(self):
        lastNode = self.head
        prevNode = None
        while lastNode.next is not None:
            prevNode = lastNode
            lastNode = lastNode.next
        prevNode.next = None
        self.size -= 1
        return lastNode
        
    def addFirst(self, item):
        new = Node(item)
        x = self.head
        new.next = x
        self.head = new
        self.size +=1
        
    def removeFirst(self):
        res = self.head.item
        x = self.head.next
        self.head = x
        self.size -= 1
        return res
    
    
    def insertAt(self, val, position):
        if position > self.size or position == -1:
            self.addLast(val)
        elif position == 0:
            self.addFirst(val)
        else:
            NewNode = Node(val)
            iterator = 0
            currentNode = self.head
            while True:
                if iterator == position:
                    NewNode.next = currentNode
                    prevnode.next = NewNode
                    break
                prevnode = currentNode
                currentNode = currentNode.next
                iterator +=1
        self.size += 1
    
    def delAt(self, position):
        if position > self.size or position == -1:
            self.size -= 1
            return self.delLast()
        elif position == 0:
            self.size -= 1
            return self.removeFirst()
        else:
            res = None
            iterator = 0
            prevnode = None
            currentNode = self.head
            while True:
                if iterator == position:
                    res = currentNode
                    prevnode.next = currentNode.next
                    break
                prevnode = currentNode
                currentNode = currentNode.next
                iterator +=1
            self.size -= 1
            return res
    
    
    def __str__(self):
        print_val = self.head
        ret = []
        while print_val != None:
            ret.append(print_val.item)
            print_val = print_val.next
        return str(ret)
```
