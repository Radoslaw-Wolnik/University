```python
class Stack:
    def __init__(self):
        self.start()
    
    def start(self):
        self.structure = []
            
    def add(self, val):
        assert isinstance(val, (str, int, float)), "This element can't be in Stack\nU can add str, int or float"
        self.structure.append(val)
    
    def get(self):
        if self.is_empty():
            return 'Stack is empty'
        return self.structure.pop()
    
    def peak(self):
        if self.is_empty():
            return 'Stack is empty'
        return self.structure[-1]
            
    def __str__(self):
        return str(self.structure)
    
    def is_empty(self):
        if len(self.structure) == 0:
            return True
        else:
            return False
```

```python
def count_postfix(exp):
    operators = ['+', '-', '*', '/']
    Res = Stack()
    exp = exp.split()
    
    for el in exp:
        if el in operators and Res.is_empty() == False:
            b = Res.get()
            a = Res.get()
            if el == '+':
                a = a + b
            elif el == '-':
                a = a - b
            elif el == '/':
                a = a/b
            elif el == '*':
                a = a*b
            Res.add(a)
        
        else:
            Res.add(float(el))
            
    return Res.get()
    
    
x = "20 10 + 75 45 - *"
count_postfix(x)
```

```python
from copy import deepcopy

class Kolejka:
    # self.data structure
    # self.len lenght of structure
    # self.size number of elements in structure

    def __init__(self, lenght = 10):
        self.set_len(lenght) # 10 = default size
        self.MakeNull()
        self.front = 0
        
    def set_len(self, val):
        assert isinstance(val, int), 'Lenght of structure have to be an intiger value'
        self.len = val
        
    def MakeNull(self):
        self.data = [None] *self.len
        self.size = 0
        
    def Front(self):
        return self.data[self.front]
    
    def Empty(self):
        if sum([1 for el in self.data if el != None]) == 0:
            return True
        return False
    
    def Enqueue(self, val): # adds element to structure; add at the end
        position = (self.front + self.size) % self.len
        if self.data[position] != None:
            self.Resize()
            position = (self.front + self.size) % self.len
        self.data[position] = val
        self.size += 1
        
    def Dequeue(self): # returns element from structure; returns the element that was on structure longest time
        val = self.data[self.front]
        self.data[self.front] = None
        self.front = (self.front + 1) % self.len
        self.size -= 1
        return val
    
    def Resize(self):
        temp = deepcopy(self)
        self.set_len(self.len *2)
        self.MakeNull()
        while temp.Empty() == False:
            self.Enqueue(temp.Dequeue())
            
    def Resize2(self):
        temp = self.data
        walk = self.Front()
        self.set_len(self.len *2)
        self.MakeNull()
        for i in range(len(temp)):
            self.Enqueue(temp[walk])
            walk = (walk+1) % len(temp)
        
        
    def __str__(self):
        return f'front: {self.front} size: {self.size} lenght: {self.len}\n' + f'{self.data}'
```

```python
class Stack:
    def __init__(self):
        self.start()
    
    def start(self):
        self.structure = []
            
    def push(self, val):
        assert isinstance(val, (str, int, float)), "This element can't be in Stack\nU can add str, int or float"
        self.structure.append(val)
    
    def pop(self):
        if self.is_empty():
            return 'Stack is empty'
        return self.structure.pop()
    
    def peak(self):
        if self.is_empty():
            return 'Stack is empty'
        return self.structure[-1]
            
    def __str__(self):
        return str(self.structure)
    
    def is_empty(self):
        if len(self.structure) == 0:
            return True
        else:
            return False
```

```python
class Kolejka_2_stosy():
    def __init__(self):
        self.One = Stack()
        self.Two = Stack()
    
    def enqueue(self, val):
        self.One.push(val)
        
    def dequeue(self):
        res = None
        while self.One.peak() != 'Stack is empty':
            self.Two.push(self.One.pop())
        res = self.Two.pop()
        while self.Two.peak() != 'Stack is empty':
            self.One.push(self.Two.pop())
        return res
    
Kolejka = Kolejka_2_stosy()
Kolejka.enqueue('E')
Kolejka.enqueue(1)
Kolejka.enqueue(2)
Kolejka.enqueue(3)
print(Kolejka.dequeue())
print(Kolejka.dequeue())
print(Kolejka.dequeue())
```
