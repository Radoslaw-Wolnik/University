Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python
def sort_bubble(L):
    '''funkcja sortuje liste L za pomoca aglgorytmu bubble sort
    funkcja przyjmuje liste
    jesli do funkcji przekazana będzie wartość Type != list to funkcja funkcja zwraca 'Not list'
    jesli do funkcji przekazana będzie wartość któej elementy nie mogą być porównane lub zamienione to funkcja
    zwróci 'Wrong type of element in list'
    jeśli zostanie przekazany dobry typ funkcja zwróci posortowana liste
    '''
    
    if type(L) != list:
        return 'Not list'
    for el in L:
        if not ( isinstance(el, (int, float, str)) ):
            return 'Wrong type of element in list'
    
    # ------------------------------------------------------

    sortowanie = True
    licznik = 1
    while sortowanie:
        sortowanie = False
        for i in range(len(L)-licznik):
            if L[i+1] < L[i]:
                L[i+1], L[i] = L[i], L[i+1]
                sortowanie = True
        licznik += 1

    return L
        
        
lista = [5, 6, -1, 0, 4, 7, 2, 3]

print(sort_bubble(lista))
print(sort_bubble([i for i in range (10)]))
print(sort_bubble([i for i in range (10,0,-1)]))
print(sort_bubble(['v', 'd', 's', 'h', 'a', 'i']))
print(sort_bubble('vdshai'))
print(sort_bubble(['vd','aa','ccc']))
print(sort_bubble([[1],[2]]))
```

## Zad 2. Sortowanie przez wstawianie.
Napisz funkcję realizującą sortowanie przez wstawianie.

Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python

def wstawienie_sort(L):
    '''funkcja sortuje liste za pomoca algorytmu wstawiania
    przyjmuje liste
    jesli do funkcji przekazana będzie wartość Type != list to funkcja zwraca 'Not list'
    jesli do funkcji przekazana będzie wartość któej elementy nie mogą być porównane lub zamienione to funkcja
    zwróci 'Wrong type of element in list'
    jesli parametr jest właściwy zwraca posortowana liste'''
    
    if type(L) != list:
        return 'Not list'
    for el in L:
        if not ( isinstance(el, (int, float, str)) ):
            return 'Wrong type of element in list'
        
        
    for marker in range(1,len(L)):
        for i in range (marker-1,-1,-1):
            if L[i] > L[marker]:
                L[i], L[marker] = L[marker], L[i]
                marker -= 1
            else:
                break
    return L
                
                
print(wstawienie_sort([5, 6, -1, 0, 4, 7, 2, 3, 102, 88]))
print(wstawienie_sort([i for i in range (10)]))
print(wstawienie_sort([i for i in range (10,0,-1)]))

print(wstawienie_sort('geografia'))
print(wstawienie_sort([[1],[2]]))
```

## Zad 3. Sortowanie przez scalanie
Napisz funkcję realizującą sortowanie przez scalanie.

Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python
# pesymistycznie O(nlogn)
#
# optymistycznie O(nlogn)

def merge(left, right):
    result = []
    #print(left)
    #print(right)
    while len(left) > 0 and len(right) > 0:
        if left[0] > right[0]:
            result.append(right[0])
            del right[0]
        else:
            result.append(left[0])
            del left[0]
   
    result += left + right 
    #print(result)
    return result


def merge_sort(L):
    '''funkcja sortuje liste za pomoca algorytmu scalenia
    przyjmuje liste
    jesli do funkcji przekazana będzie wartość Type != list to funkcja zwraca 'Not list'
    jesli do funkcji przekazana będzie wartość któej elementy nie mogą być porównane lub zamienione to funkcja
    zwróci 'Wrong type of element in list'
    jesli parametr jest właściwy zwraca posortowana liste'''
    
    if type(L) != list:
        return 'Not list'
    for el in L:
        if not ( isinstance(el, (int, float, str)) ):
            return 'Wrong type of element in list'
    
    if len(L) > 1:
        mid = len(L)//2
        left = L[:mid]
        right = L[mid:]
        left = merge_sort(left)
        right = merge_sort(right)
        return merge(left, right)
    else:
        return L

print(merge_sort([3,5,-1,2,7,4]))
print(merge_sort([-3,4,5,7,2,4,-8,0,-3]))
print(merge_sort([5, 6, -1, 0, 4, 7, 2, 3, 102, 88]))
print(merge_sort([i for i in range (10)]))
print(merge_sort([i for i in range (10,0,-1)]))

print(merge_sort('geografia'))
print(merge_sort([[1],[2]]))
```

## Zad 4. Sortowanie przez zliczanie
Napisz funkcję realizującą sortowanie przez zliczanie.

Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python


def zliczenie(L):
    '''funkcja sortuje liste za pomoca algorytmu zliczenia
    przyjmuje liste
    jesli do funkcji przekazana będzie wartość Type != list to funkcja zwraca 'Not list'
    jesli do funkcji przekazana będzie wartość któej elementy nie mogą być porównane lub zamienione to funkcja
    zwróci 'Wrong type of element in list'
    jesli parametr jest właściwy zwraca posortowana liste'''
    
    if type(L) != list:
        return 'Not list'
    for el in L:
        if not ( isinstance(el, (int, float, str)) ):
            return 'Wrong type of element in list'
    
    min = L[0]
    max = L[0]
    
    for el in L:
        if el > max:
            max = el
        elif el < min:
            min = el
    
    histogram = {i : 0 for i in range (min,max+1,1)}
    
    for el in L:
        histogram[el] += 1
    
    result = []
    for key, val in histogram.items():
        for i in range(val):
            result.append(key)
        
    return result

print(zliczenie([-3,4,5,7,2,4,-8,0,-3]))
print(zliczenie([5, 6, -1, 0, 4, 7, 2, 3, 102, 88]))
print(zliczenie([i for i in range (10)]))
print(zliczenie([i for i in range (10,0,-1)]))

print(zliczenie('geografia'))
print(zliczenie([[1],[2]]))
```

## Zad 5. Sortowanie głupie
Jest to przykład złego algorytmu sortującego, które jednak daje zawsze poprawne wyniki. Zasada działania jest bardzo prosta. Przeglądamy kolejne pary sąsiednich elementów sortowanego zbioru. Jeśli bieżąco przeglądana para elementów jest w złej kolejności, elementy pary zamieniamy miejscami i całą operację rozpoczynamy od początku zbioru. Jeśli przeglądniemy wszystkie pary, zbiór będzie posortowany.

Jeżeli zbiór danych `d` ma długość `n`, to lista kroków jest nastepująca

    K01:	Dla i = 0, 1, 2, ..., n - 1: wykonuj K02...K04
    K02:	    Jeśli d[i] ≤ d[i + 1], to wykonaj następny obieg K01
    K03:	    Zamień miejscami d[i] oraz d[i + 1]
    K04:	    Idź do K01
    K05:	Zakończ

Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python


def stupid_sort(L):
    '''funkcja sortuje liste za pomoca algorytmu sortowania glupiego
    przyjmuje liste
    jesli do funkcji przekazana będzie wartość Type != list to funkcja zwraca 'Not list'
    jesli do funkcji przekazana będzie wartość któej elementy nie mogą być porównane lub zamienione to funkcja
    zwróci 'Wrong type of element in list'
    jesli parametr jest właściwy zwraca posortowana liste'''
    
    if type(L) != list:
        return 'Not list'
    for el in L:
        if not ( isinstance(el, (int, float, str)) ):
            return 'Wrong type of element in list'
    
    notSorted = True
    while notSorted:
        notSorted = False
        for i in range(len(L)-1):
            if L[i] > L[i+1]:
                L[i], L[i+1] = L[i+1], L[i]
                notSorted = True
                break
    return L
                
                
print(stupid_sort([5, 6, -1, 0, 4, 7, 2, 3]))
print(stupid_sort([i for i in range (10)]))
print(stupid_sort([i for i in range (10,0,-1)]))
print(stupid_sort([i for i in range (5,0,-1)]))

print(stupid_sort('elephant'))
print(stupid_sort([[1],[2],(4)]))
```

## Zad 6. Bogosort
Napisz funkcję realizującą [bogosort](https://pl.wikipedia.org/wiki/Bogosort).

Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python
from random import shuffle



def bogosort(L):
    '''funkcja sortuje liste za pomoca algorytmu bogosort
    przyjmuje liste
    jesli do funkcji przekazana będzie wartość Type != list to funkcja zwraca 'Not list'
    jesli do funkcji przekazana będzie wartość któej elementy nie mogą być porównane lub zamienione to funkcja
    zwróci 'Wrong type of element in list'
    jesli parametr jest właściwy zwraca posortowana liste'''
    
    if type(L) != list:
        return 'Not list'
    for el in L:
        if not ( isinstance(el, (int, float, str)) ):
            return 'Wrong type of element in list'
    
    notSorted = True
    n = 0
    while notSorted:
        notSorted = False
        for i in range(len(L)-1):
            if L[i] > L[i+1]:
                notSorted = True
                shuffle(L)
                break
        n += 1
        #print('still working', n)  # pokazuje że coś się dzieje
    return L



print(bogosort([i for i in range (10)]))
print(bogosort([5, 6, -1, 0, 4, 12]))

print(bogosort([[2],['a'],[1]]))
print(bogosort('Hello world'))
```

```python
from random import randint
# pesymistycznie n^2 chyba
# średnio nlogn chyba
# optymistycznie n

def quicksort_r(L, pivotPoint = 'a', plus = False):
    '''funkcja sortuje tablicę rekurencyjnie za pomocą algorytmu quicksort
    L = lista do posortowania
    pivotPoint = (a) początku, (b) końcu, (c) w środku lub w (d) losowym miejscu listy
    plus = True, jeśli chcemy umożliwić sortowanie krótkich fragmentów inaczej niż quicsort
    plus = False, całość sortowania za pomocą quicksort
    
    jeśli type(L) != list zwraca 'not list'
    jeśli elementy w liście nie mogą być sortowane funkcja zwraca 'Wrong type of element in list'
    jeśli pivotPoint jest różny od 'a', 'b', 'c' lub 'd' zwraca 'Wrong parametr pivotPoint (it should be 'a', 'b', 'c' or 'd')'
    jęsli type(plus) != bool zwraca 'Wrong type of parametr plus (it should be True or False)'
    
    w przypadku wprowadzenia poprawnych parametrow zwraca posortowaną listę'''
    
    if type(L) != list:
        return 'Not list'
    for el in L:
        if not ( isinstance(el, (int, float, str)) ):
            return 'Wrong type of element in list'
    if not (pivotPoint in ['a','b','c','d']):
        return "Wrong parametr pivot point (it should be 'a', 'b', 'c' or 'd')"
    if type(plus) != bool:
        return 'Wrong type of parametr plus (it should be True or False)'
    
    
    until = 1
    if plus == True:
        until = 5

    less = []
    equal = []
    greater = []

    if len(L) > until:
        pivotdict = {'a' : 0, 'b' : len(L)-1, 'c' : len(L)//2, 'd' : randint(0,len(L)-1)}
        pivot = L[pivotdict[pivotPoint]]
#        print(pivot)
        for el in L:
            if el < pivot:
                less.append(el)
            elif el == pivot:
                equal.append(el)
            elif el > pivot:
                greater.append(el)
        return quicksort_r(less, pivotPoint, plus) + equal + quicksort_r(greater, pivotPoint, plus)
    
    elif plus == True:
#        print('bubble')
        sortowanie = True
        licznik = 1
        while sortowanie:
            sortowanie = False
            for i in range(len(L)-licznik):
                if L[i+1] < L[i]:
                    L[i+1], L[i] = L[i], L[i+1]
                    sortowanie = True
            licznik += 1
        return L
    else:  
        return L
    

print(quicksort_r([5, 6, -1, 0, 4, 7, 2, 3, 102, 88], 'a', False))
print(quicksort_r([5, 6, -1, 0, 4, 7, 2, 3, 102, 88], 'b', False))
print(quicksort_r([randint(1,100) for i in range (10)], 'c', False))
print(quicksort_r([randint(1,100) for i in range (20)], 'd', True))


# błędy
print(quicksort_r([[1],[2]]))
print(quicksort_r([5, 6, -1, 0, 4, 7, 2, 3, 102, 88], 1))
print(quicksort_r([5, 6, -1, 0, 4, 7, 2, 3, 102, 88], 'a', 1))
```
