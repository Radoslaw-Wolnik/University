```python
def wstawienie_sort(L):
     
    for marker in range(1,len(L)):
        for i in range (marker-1,-1,-1):
            if L[i] > L[marker]:
                L[i], L[marker] = L[marker], L[i]
                marker -= 1
            else:
                break
    return L
```

```python
def sort_bubble(L):

    sortowanie = True
    licznik = 1
    while sortowanie:
        sortowanie = False
        for i in range(len(L)-licznik):
            if L[i+1] < L[i]:
                L[i+1], L[i] = L[i], L[i+1]
                sortowanie = True
        licznik += 1

    return L
        

create_plt(sort_bubble)
```

```python
# zadanie 6
# search smallest than take it to the beggining

# optymistic O(n^2)
# average O(n^2)
# pesymistic O(n^2)

def selection_sort(L):
    for i in range(len(L)):
        min = i
        for j in range(i+1,len(L)):
            if L[j] < L[min]:
                min = j
        L[i], L[min] = L[min], L[i]
        return L
```

```python
def merge(left, right):
    result = []
    #print(left)
    #print(right)
    while len(left) > 0 and len(right) > 0:
        if left[0] > right[0]:
            result.append(right[0])
            del right[0]
        else:
            result.append(left[0])
            del left[0]
   
    result += left + right 
    #print(result)
    return result


def merge_sort(L):
    '''funkcja sortuje liste za pomoca algorytmu scalenia
    przyjmuje liste'''
    
    
    if len(L) > 1:
        mid = len(L)//2
        left = L[:mid]
        right = L[mid:]
        left = merge_sort(left)
        right = merge_sort(right)
        return merge(left, right)
    else:
        return L
```

```python
# quick sort recursive
from random import randint
# pesymistycznie n^2 chyba
# średnio nlogn chyba
# optymistycznie n

def quicksort_r(L, pivotPoint = 'a'):
    '''funkcja sortuje tablicę rekurencyjnie za pomocą algorytmu quicksort
    L = lista do posortowania
    pivotPoint = (a) początku, (b) końcu, (c) w środku lub w (d) losowym miejscu listy'''

    less = []
    equal = []
    greater = []

    if len(L) > 1:
        pivotdict = {'a' : 0, 'b' : len(L)-1, 'c' : len(L)//2, 'd' : randint(0,len(L)-1)}
        pivot = L[pivotdict[pivotPoint]]
#       print(pivot)
        for el in L:
            if el < pivot:
                less.append(el)
            elif el == pivot:
                equal.append(el)
            elif el > pivot:
                greater.append(el)
        return quicksort_r(less, pivotPoint) + equal + quicksort_r(greater, pivotPoint)
    
    else:
        return L
```

```python
import time
import matplotlib.pyplot as plt
from random import choice


def create_plt(sorting_algorytm):
    '''jako algorytm funkcji przekazywana jest odpowiednia funkcja sortujaca'''
    N = [10, 100, 300, 500, 700]
    #N =[2,3,4,10]
    
    tries = 10
    optymistic_plt = []
    pesymistic_plt = []
    average_plt = []
        
    for lenght in N:
        
        optymistyczny = [el for el in range(lenght)]
        pesymistyczny = [lenght - el - 1 for el in range(lenght)]
        sredni = [choice(optymistyczny) for _ in range(lenght)]
        
        optymistic_avr = 0
        pesymistic_avr = 0
        average_avr = 0
        for _ in range(tries):
            start = time.time()
            sorting_algorytm(optymistyczny)
            end = time.time()
            optymistic_avr += end - start
            
            start = time.time()
            sorting_algorytm(pesymistyczny)
            end = time.time()
            pesymistic_avr += end - start
            
            start = time.time()
            sorting_algorytm(sredni)
            end = time.time()
            average_avr += end - start
            
        optymistic_plt.append(optymistic_avr/tries)
        pesymistic_plt.append(pesymistic_avr/tries)
        average_plt.append(average_avr/tries)
        
    plt.figure(figsize=(8, 5))
    plt.plot(N, optymistic_plt, '-', color  = 'skyblue', label = 'optymistyczna sytuacja')
    plt.plot(N, pesymistic_plt, '-', color  = 'crimson', label = 'pesymistyczna sytuacja')
    plt.plot(N, average_plt, '-', color  = 'green', label = 'srednia sytuacja')
    plt.legend(bbox_to_anchor=(1.04,0.5), loc='center left', shadow=True, fontsize='x-large')
    plt.xlabel('dlugosc listy')
    plt.ylabel('czas [s]')
    plt.title(sorting_algorytm.__name__)
    plt.show()
        
create_plt(wstawienie_sort)
```

```python
# wyszukiwanie max min jednocześnie

def min_max_it(L):
    '''funkcja wyznacza minimalną i maksymalną wartość w liście
    przyjmuje liste
    zwraca słownik: {'min' : min, 'max' : max}
    zlozonosc obliczeniowa: 3/2 * n - 2 dla parzystych len(L)
    zlozonosc obliczeniowa: 3/2 * n - 3/2 dla nieparzystych len(L)'''
    i = 0
    if len(L)%2 == 1:
        i = 1
    min = L[0]
    max = L[0]
    while i < len(L):
        if L[i] > L[i+1]:
            if L[i] > max:
                max = L[i]
            if L[i+1] < min:
                min = L[i+1]
        else:
            if L[i] < min:
                min = L[i]
            if L[i+1] > max:
                max = L[i+1]
        i+=2
    return {'min' : min, 'max' : max}

min_max_it([2,3,7,1,-1,2,87,23,54,-23,-43])
```

```python
# binary search
# optymistyczna O(1)
# srednia O(logn)
# pesymistyczna O(logn)

def binary_search(L, search):
    low = 0
    high = len(L)
    idx = 0
    while len(L) > 1:
        mid = len(L)//2
        if search == L[mid]:
            return idx + mid
        if search < L[mid]:
            L = L[low:mid]
            continue
        if search > L[mid]:
            L = L[mid:high]
            idx += mid
            
    if L[0] != search:
        return None
    return idx
```

```python
def linear_serach(L, search):
    for i in range(len(L)):
        if L[i] == search:
            return i
    return None
```

```python
def wartownik_linear_search(L, search):
    L.append(-1)
    for i in range(len(L)):
        if L[i] == search:
            break
            
    if i == len(L)-1:
        return L[i]
    return i
```

```python
def Ackerman_function(m,n):
    if m == 0:
        return n + 1
    if n == 0:
        return Ackerman_function(m-1,1)
    return Ackerman_function(m-1, Ackerman_function(m, n-1))

print(Ackerman_function(2,1))  # 5
print(Ackerman_function(3,2))  # 29
print(Ackerman_function(3,4))  # 125
# https://en.wikipedia.org/wiki/Ackermann_function#Table_of_values
```
