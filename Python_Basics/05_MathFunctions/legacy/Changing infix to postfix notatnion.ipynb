```python
# Zadanie 1
class Stack:
    def __init__(self, kind):
        self.set_kind(kind)
        self.start()
    
    def start(self):
        self.structure = []
    
    def set_kind(self, kind):
        assert kind in ['beginning', 'ending'], f'There is no such a kind of stack as {kind}\nChose from: ["beginning", "ending"]'
        if kind == 'beginning':
            self.kind = False
        if kind == 'ending':
            self.kind = True
        
    def get_kind(self):
        return self.kind
            
    def add(self, val):
        assert isinstance(val, (str, int, float)), "This element can't be in Stack\nU can add str, int or float"
        if self.get_kind() == True:
            self.structure.append(val)
        else:
            self.structure.insert(0, val)
    
    def get(self):
        if self.structure == []:
            return 'Stack is empty'
        if self.get_kind == True:
            return self.structure.pop(-1)
        else:
            return self.structure.pop(0)
            
    def __str__(self):
        return str(self.structure)
    
    
    
Stack01 = Stack('ending')
for el in [1,2,3,4,'Ania','Kasia','Jola',-1,5]:
    Stack01.add(el)
print(Stack01)
for i in range(10):
    print(Stack01.get())
    
print()
print('-'*10)
print('-'*10)
print()

Stack02 = Stack('beginning')
for el in [1,2,3,4,'Ania','Kasia','Jola',-1,5]:
    Stack02.add(el)
print(Stack02)
for i in range(10):
    print(Stack02.get())

print('\n' + '-'*10 + '\n')


# wrong input
#Stack03 = Stack('mandarynka')
#Stack03 = Stack('beginning')
#Stack03.add(('Kasia', 1))
```
