### Zadanie 1 Wektor [40]
Zaimplementuj klasę wektor:
(dodawanie, mnożenie przez skalar, odejmowanie, wyznaczenie długości, iloczyn skalarny wektorów, czy wektory sa równe, czy wektory są prostopadłe, czy wektory są równoległe)

```python
class Vector():
    '''
    Klasa Vector
    atrybuty:
        x - wspolzedna x
        y - wspolzedna y
        z - wspolzedna z
    funkcje:
        set_vector - ustawia x, y, z
        check_xyz - sprawdza czy x, y, z, spelniaja warunki
        add - dodaje do wektora inny wektor
        substract - odejmuje od wektora inny wektor
        multiply_skalar - mnozy wektor przez skalar
        lenght - zwraca dlugosc wektora
        iloczyn_skalarny - zwraca wartosc iloczynu skalarnego wektora i podanego do funkcji wektora
        iloczyn_wektorowy - zwraca wektor bedacy wynikiem iloczynu wektorowego wektora i podanego do funkcji wektora
        equal_to - zwraca True/False czy podany do funkcji wektor jest rowny wektorowi w klasie
        prostopadle - zwraca True/False czy wektor podany do funkcji jest prostopadly do wektora w klasie
        rownolegle - zwraca True/False czy wektor podany do funkcji jest rownolegly do wektora w klasie
        __str__ - zwraca str: 'vector: [x, y, z]'
    '''
    
    def __init__(self, x, y, z):
        self.set_vector(x,y,z)
        
    def set_vector(self, x, y, z):
        assert isinstance(x, (int, float)), 'you can only input int or double value for x'
        self.x = x
        assert isinstance(y, (int, float)), 'you can only input int or double value for y'
        self.y = y
        assert isinstance(z, (int, float)), 'you can only input int or double value for z'
        self.z = z
        
    def check_xyz(self, x, y, z):
        if isinstance(x, (int, float)) and isinstance(y, (int, float)) and isinstance(z, (int, float)):
            return True
        return False
        
    # ----- operacje -----
        
    def add(self, x, y, z):
        if self.check_xyz(x, y, z):
            self.x += x
            self.y += y
            self.z += z
        else:
            print('all input have to be int or double')
        
    def substract(self, x, y, z):
        if self.check_xyz(x, y, z):
            self.x -= x
            self.y -= y
            self.z -= z
        
    def multiply_skalar(self, num):
        assert isinstance(num, (int, float)), 'you can only input int or double value for number'
        self.x *= num
        self.y *= num
        self.z *= num
        
    def lenght(self):
        result = self.x**2 + self.y**2 + self.z**2
        return result**(1/2)
    
    def iloczyn_skalarny (self, x, y, z):
        if self.check_xyz(x, y, z):
            result = 0
            result += self.x * x
            result += self.y * y
            result += self.z * z
            return result
        else:
            return 'all input have to be int or double'
        
    def iloczyn_wektorowy (self, a, b, c):
        if self.check_xyz(a, b, c):
            result = []
            result.append(self.y*c - self.z*b)
            result.append(self.z*a - self.x*c)
            result.append(self.x*b - self.y*a)
            return result
        else:
            return 'all input have to be int or double'
    
    def equal_to(self, x, y, z):
        if self.check_xyz(x, y, z):
            result = []
            if self.x == x:
                result.append(1)
            if self.y == y:
                result.append(1)
            if self.z == z:
                result.append(1)
            if sum(result) == 3:
                result = True
            else:
                result = False
            return result
        else:
            return 'all input have to be int or double'
    
    def prostopadle(self, x, y, z):
        # iloczyn skalarny = 0
        if self.check_xyz(x, y, z):
            if self.iloczyn_skalarny(x,y,z) == 0:
                return True
            return False
        else:
            return 'all input have to be int or double'
    
    def rownolegle(self, x, y, z):
        # jesli iloczyn wektorowy = [0,0,0]
        if self.check_xyz(x, y, z):
            if self.iloczyn_wektorowy(x, y, z) == [0,0,0]:
                return True
            return False
        else:
            return 'all input have to be int or double'
        
        
    def __str__(self):
        return 'vector: ' + str([self.x, self.y, self.z])
        
        
k = Vector(1,2,3)
print(k)
k.add(1,1,1)
print(k)
k.substract(1,1,1)
print(k)
k.multiply_skalar(3)
print(k)
print(k.lenght())
print(k.iloczyn_skalarny(2,2,2))
print(k)
print(k.iloczyn_wektorowy(1,2,3))
print(k.equal_to(1,1,1))
print(k.equal_to(3,6,9))
print(k.rownolegle(1,2,3))
print(k.rownolegle(2,1,3))
print(k.prostopadle(1,2,3))
print(k.prostopadle(-3,0,1))
```

jesli gracz swoim ruchem spowoduje kolizje (bedzie na tym samym polu co drugi gracz) to gra sie konczy i przegrywa.

```python
class Board():
    '''
    Klasa Board
    atrybuty:
        rockets = {rocket_object : [x, y], rocket_object2 : [x, y]}
    funkcje:
        set_place - ustawia miejsce rakiety
        get_place - dostaje miejsce rakiety
        __str__ - rysuje plasze z rakietami
        possibilities - zwraca dostępne dla gracza mozliwosci ruchu
        move - przesuwa rakiete
        refuel - zwieksza ilosc paliwa 0 zadana wartosc
        attack - atakuje drugą rakiete
        '''
    def __init__(self, rocket1, rocket2):
        self.set_rockets(rocket1, 5, 2, rocket2, 5, 7)
            
    def set_rockets(self, rocket1, x1, y1, rocket2, x2, y2):
        self.rockets = {}
        if type(rocket1) == Rocket:
            self.rockets[rocket1] = [x1, y1]
        else:
            print('zly objekt')
        if type(rocket2) == Rocket:
            self.rockets[rocket2] = [x2, y2]
        else:
            print('zly objekt')
            
    def set_place(self, rocket, x, y):
        self.rockets[rocket] = [x,y]
            
            
    def get_place(self, rocket):
        return self.rockets[rocket]
    
    def __str__(self):
        result = []
        place = []
        for key, val in self.rockets.items():
            place.append((key, val))
        for i in range(10):
            for j in range(10):
                if [i, j] == place[0][1]:
                    result.append('|' + place[0][0].get_letter())
                    continue
                if [i, j] == place[1][1]:
                    result.append('|' + place[1][0].get_letter())
                    continue
                result.append('|.')
            result.append('|\n')
        return ''.join(result)
    
    
    
    def possibilites(self, rocket):
        self.possible = {'move': False, 'wait' : False, 'refuel': False, 'attack': False}
        place = self.get_place(rocket)
        x, y = place[0], place[1]
        result = ''
        
        if rocket.get_energy() > 0:
            result = result + 'move'
            self.possible['move'] = True
        if rocket.get_energy() == 0:
            result = result + 'wait'
            self.possible['wait'] = True
            
        if y == 2 or y == 7:
            result = result + ', refuel'
            self.possible['refuel'] = True
            
        if rocket.get_attack() == True:
            druga = []
            for key in self.rockets.keys():
                if rocket != key:
                    druga = self.get_place(key)
            if druga[0]+2 == x or druga[0]-2 == x or druga[1]-2 == y or druga[1]+2 == y:
                result = result + ', attack'
                self.possible['attack'] = True
            if (druga[0]+1 == x and druga[1]+1 == y) or (druga[0]-1 == x and druga[1]+1 == y) or (druga[0]+1 == x and druga[1]-1 == y) or (druga[0]-1 == x and druga[1]-1 == y):
                result = result + ', attack'
                self.possible['attack'] = True
        return result
    
    
    def move(self, rocket, direction, num):
        if self.possible['move'] == True:
            try: 
                num = int(num)
            except:
                print('you can only move by intiger value')
                return None
            temp = self.get_place(rocket)
            place = temp
            if num > rocket.get_energy():
                num = rocket.get_energy()
            rocket.set_energy(rocket.get_energy() - num)
            if num >= 10:
                num = num//10
            if direction == 'left':
                place[1] -= num
                if place[1] < 0:
                    place[1] = 10 + place[1]

            if direction == 'right':
                place[1] += num
                if place[1] > 9:
                    place[1] = place[1] - 10

            if direction == 'down':
                place[0] += num
                if place[0] > 9:
                    place[0] = place[0] - 10

            if direction == 'up':
                place[0] -= num
                if place[0] < 0:
                    place[0] = place[0] + 10

            kolizja = False
            self.set_place(rocket, place[0], place[1])
            for key in self.rockets.keys():
                if rocket != key:
                    key.set_attack(True)
                    if self.rockets[key] == self.rockets[rocket]:
                        kolizja = True
            return kolizja
    
    def refuel(self, rocket, num):
        if self.possible['refuel'] == True:
            rocket.set_energy(rocket.get_energy() + num)
        for key in self.rockets.keys():
            if rocket != key:
                key.set_attack(True)
    
    def attack(self, rocket):
        if self.possible['attack'] == True:
            rocket.set_life(rocket.get_life()-1)
            rocket.set_attack(False)
    
    
    
        
    
class Rocket():
    '''
    Klasa Rocket
    atrybuty:
        name - nazwa
        letter - litera ktora bedzie na planszy
        life - ilosc zyc
        energy - ilosc paliwa
        can_attack - czy moze atakowac (nie moze jesli zostal zaatakowany w poprzednim ruchu)
    funkcje:
        set_name - ustawia nazwe
        get_name - zwraca nazwe
        set_life - ustawia ilosc zyc
        get_life - zwraca ilosc zyc
        set_letter - ustawia litere ktora reprezentuje gracza na planszy
        get_letter - zwraca litere
        set_energy - ustawia ilosc paliwa
        get_energy - zwraca ilosc paliwa
        set_attack - ustawia czy gracz moze atakowac
        get_attack - zwraca czy gracz moze atakowac
        __str__ - zwraca str: 'Gracz: ' + name + '\nInicjal: ' + letter + '\nLiczba zyc: ' +  liczba_zyc + '\n
                               Ilosc energii: ' + ilosc_energii + '\n'
'''
    def __init__(self, name, letter):
        self.set_name(name)
        self.set_letter(letter)
        self.set_life(3)
        self.set_energy(10)
        self.set_attack(True)
        
    def set_name(self, name):
        if type(name) == str:
            self.name = name
        else:
            print('wrong type of name')
    
    def get_name(self):
        return self.name
    
    def set_letter(self, letter):
        assert isinstance(letter, str), 'Initial should be str'
        assert len(letter) == 1, 'Initial should have 1 character'
        if (letter >= 'a' and letter <= 'z') or (letter >= 'A' and letter <= 'Z'):
            self.letter = letter
        else:
            print('letter should be a letter')
            
    def get_letter(self):
        return self.letter
    
    def set_life(self, num):
        assert isinstance(num, int), 'num of lifes should be an int val'
        self.life = num
        
    def get_life(self):
        return self.life
    
    def set_energy(self, num):
        assert isinstance(num, int), 'num of energy should be an int val'
        self.energy = num
    
    def get_energy(self):
        return self.energy
    
    def set_attack(self, boolan):
        self.can_attack = boolan
        
    def get_attack(self):
        return self.can_attack
    
    def __str__(self):
        result = 'Gracz: ' + self.get_name() + '\nInicjal: ' + self.get_letter() + '\nLiczba zyc: ' +  str(self.get_life()) + '\nIlosc energii: ' + str(self.get_energy()) + '\n'
        return result
    


zasady = '''Gracze strerują dwoma rakietami. Znajdują się na plaszy 10x10, jesli gracz do jedzie do krawedzi planszy to moze przeskoczyc na przeciwlegly kraniec. Kazdy z graczy rozpoczyna gre z 10 energii i 3 zyciami
W 3 i 7 koluminie gracze moga doladowywac energie (10)
gracze moga atakowac siebie nawzajem jesli znajduja sie jedno pole po przekatnych lub 2 pola na boki, w gore lub w dol
po zaatakowaniu gracz ktory stracil 1 zycie nie moze zaatakowac ponownie przez 1 ruch
jesli gracz swoim ruchem spowoduje kolizje (bedzie na tym samym polu co drugi gracz) to gra sie konczy i przegrywa.'''

print(zasady)
print('Gracz 1')
name = input('Podaj nazwe: ')
inicjal = input('Podaj inicjal: ')
rakieta1 = Rocket(name, inicjal)

print('Gracz 2')
name = input('Podaj nazwe: ')
inicjal = input('Podaj inicjal: ')
rakieta2 = Rocket(name, inicjal)

plansza = Board(rakieta1, rakieta2)

play = True
iterator = 0
while play:
    print(plansza)
    if iterator %2 == 0:
        print(rakieta1)
        print(plansza.possibilites(rakieta1))
        wybor = input('wybierz: ')
        
        if wybor == 'move':
            gdzie = input('left/right/up/down: ')
            ile = input('ile pol: ')
            temp = plansza.move(rakieta1, gdzie, ile)
            if temp == True:
                print('kolizja')
                print(f'Wygrywa gracz: {rakieta2.get_name()}')
                play = False

        if wybor == 'refuel':
            plansza.refuel(rakieta1, 10)
            
        if wybor == 'attack':
            plansza.attack(rakieta2)
            
        if wybor == 'wait':
            plansza.refuel(rakieta1, 2)
    else:
        print(rakieta2)
        print(plansza.possibilites(rakieta2))
        wybor = input('wybierz: ')
        
        if wybor == 'move':
            gdzie = input('left/right/up/down: ')
            ile = input('ile pol: ')
            temp = plansza.move(rakieta2, gdzie, ile)
            if temp == True:
                print('kolizja')
                print(f'Wygrywa gracz: {rakieta1.get_name()}')
                play = False

        if wybor == 'refuel':
            plansza.refuel(rakieta2, 10)

        if wybor == 'attack':
            plansza.attack(rakieta1)
            
        if wybor == 'wait':
            plansza.refuel(rakieta2, 2)
    iterator += 1
    if play != False:
        if rakieta1.get_life() == 0:
            print(f'Gracz {rakieta2.get_name()} wygrywa')
            play = False
        if rakieta2.get_life() == 0:
            print(f'Gracz {rakieta1.get_name()} wygrywa')
            play = False
```
