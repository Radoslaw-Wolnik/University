Twój kolega ma spory zbiór płyt muzycznych. Chce je uporządkować. Poprosił Ciebie, abyś napisał mu program, który pozwoli mu zarządzać tym zbiorem, czyli taką bazą płyt. O każdej płycie chce posiadać następujące informacje:

- tytuł płyty
- typ płyty (czy płyta DVD czy CD)
- czas trwania płyty
- spis utworów znajdujących się na każdej płycie
- spis wykonawców, którzy wykonują utwory na danej płycie
- numer płyty (identyfikator płyty nadany przez kolegę)
    
O każdym utworze kolega chce wiedzieć:

- tytuł utworu
- czas trwania utworu
- spis wykonawców, którzy wykonują dany utwór na danej płycie
- kompozytora danego utworu
- numer utworu na płycie
     
Program powinien umożliwiać:

- dodanie płyty do bazy danych
- wyświetlenie wszystkich płyt znajdujących się w bazie danych (same tytuły płyt)
- wyświetlenie szczegółowej informacji na temat płyty (tytuł płyty, typ płyty, czas trwania płyty, spis tytułów utworów znajdujących się na płycie z podanym numerem utworu na płycie)
- wyświetlenie wykonawców, którzy wykonują utwory na danej płycie
- wyświetlenie szczegółowej informacji na temat wybranego utworu z danej płyty (tytuł utworu, czas trwania utworu, spis wykonawców którzy wykonują dany utwór, kompozytora danego utworu)
- zapisać bazę do pliku (pickle)
- odczytać bazę z pliku

```python
import pickle
# ------------------------------------------------------------------------------------
class Utwor():
    '''
    Klasa Utwor
    Atrybuty:
        title
        time
        band
        composer
        number'''
    
    def __init__(self, title, time, band, composer, number):
        self.title = title
        self.time = time #sekund
        self.band = band
        self.composer = composer
        self.number = number
    
    def get_title(self):
        return self.title
    
    def get_time(self):
        return self.time
    
    def get_band(self):
        return self.band
    
    def get_composer(self):
        return self.composer
    
    def get_number(self):
        return self.number
    
    
# ------------------------------------------------------------------------------------
class Plyta():
    '''Klasa Plyta
    Atrybuty:
        disc_title
        disc_type
        number
        song_list - lista objektow typu Utwor
    funkcje:
        add_song - dodaje utwor do song_list
        get_title - zwraca tytul plyty
        get_disc_type - zwraca typ plyty
        get_songs - zwraca liste tytulow pisenek zawartych w song_list posortowanych wg numeru piosenki
        get_disc_time - zwraca sumaryczny czas trwania wszystkich utworow zawartych w song_list [s]
        get_composers - zwraca liste kompozytorow zawartych w song_list posortowanych wg numeru piosenki
        get_song_properties(tytl_piosenki) - zwraca slownik {'title' : 'title', 'time' : 'time[s]', 
            'band' : 'band', 'composer' : 'composer'}
        save_to_pickle_disc - zwraca wszystkie piosenki jako liste slownikow, gdzie kazdy slownik sklada sie
            z wszystkich informacji na temat piosenki 
            { ('disc_title', 'disc_type', 'disc_number') : 
             [ {'title' : 'name', 'time' : 60, 'band' : 'name', 'composer' : 'name', 'number' : 1}, {song2}, {song3} ] }
        '''
    
    def __init__(self, disc_title, disc_type, number):
        self.title = disc_title
        self.disc_type = disc_type
        self.number = number
        self.song_list = []
        
    def add_song(self, title, time, band, composer, number):
        self.song_list.append(Utwor(title, time, band, composer, number))
        
    def get_title(self):
        return self.title
    
    def get_disc_type(self):
        return self.disc_type
        
    def get_songs(self):
        lista_tytulow = []
        i = 1
        while i != len(self.song_list)+1:
            for song in self.song_list:
                if song.get_number() == i:
                    lista_tytulow.append(song.get_title())
                    i += 1
        return lista_tytulow
    
    def get_disc_time(self):
        disc_time = 0
        for song in self.song_list:
            disc_time += song.get_time()
        return disc_time
    
    def get_composers(self):
        composers = []
        i = 1
        while i != len(self.song_list)+1:
            for song in self.song_list:
                if song.get_number() == i:
                    composers.append(song.get_composer())
                    i += 1
        return composers
    
    def get_song_properties(self, song_title):
        result = {'title' : '', 'time' : '', 'band' : '', 'composer' : ''}
        for song in self.song_list:
            if song_title == song.get_title():
                result['title'] = song_title
                result['time'] = song.get_time()
                result['band'] = song.get_band()
                result['composer'] = song.get_composer()
                break
        return result
    
    def save_to_pickle_disc(self):
        # tworzy liste slownikow, gdzie kazdy slownik zawiera wszystkie informacje na temat piosenki
        # result = {() : [{},{},{}]}
        song_list_pickle = [] # [{},{},{}]
        for song in self.song_list:
            temp = {'title' : '', 'time' : 0, 'band' : '', 'composer' : '', 'number' : 0}
            temp['title'] = song.get_title()
            temp['time'] = song.get_time()
            temp['band'] = song.get_band()
            temp['composer'] = song.get_composer()
            temp['number'] = song.get_number()
            song_list_pickle.append(temp)
        # disc_title, disc_type, number
        disc_key = (self.title, self.disc_type, self.number)
        
        return {disc_key : song_list_pickle}


# ------------------------------------------------------------------------------------
class Baza_plyt():
    '''Klasa Baza_plyt
    Artibutes:
        lista_plyt
    funkcje:
        add_disc - pozwala na dodanie plyty (objekt typu Plyta)
        get_disc_titles - zwraca liste z tytulami plyt
        get_properties(tytul_plyty) - wypisuje informacje na temat danej plyty 
            (tytul, typ, czas trwania, utwory w odpowiedniej kolejnosci)
        get_composers_songs(tytul_plyty) - wypisuje piosenki i ich kompozytow w odpowiedniej kolejnosci
        get_song_in_disc_properties(disc_title, song_title) - wypisuje wlasiwosci konkretnej piosenki na konkretnej plycie
            (tytl, czas trwania, zespol, kompozytora)
        save_to_pickle(file_name) - zapisuje cala baze danych do pliku binarnego o podanej nazwie  
        unfreez_from_pickle(file_name) - odczytuje baze danych z pliku binarnego o podanej nazwie
        '''
    
    def __init__(self):
        self.lista_plyt = []
        
    def add_disc(self, disc): # zmienna disc typu Plyta
        self.lista_plyt.append(disc)
        
    def get_disc_titles(self):
        lista_tytulow = []
        for disc in self.lista_plyt:
            lista_tytulow.append(disc.get_title())
        return lista_tytulow
    
    def get_properties(self, title):
        for disc in self.lista_plyt:
            if title == disc.get_title():
                print('tytul plyty: ' + title)
                print('typ plyty: ' + disc.get_disc_type())
                temp = disc.get_disc_time()
                temp = str(temp//60) + 'min ' + str(temp - 60 * (temp//60)) + 's'
                print('czas trwania plyty: ' + temp)
                i = 1
                for song in disc.get_songs():
                    print(str(i) + '. ' + song) # musza byc po kolei poniewaz funkcja disc_get_songs uklada je wg numerow
                    i += 1
    
    def get_composers_songs(self, title):
        songs = []
        composers = []
        for disc in self.lista_plyt:
            if title == disc.get_title():
                songs = disc.get_songs()
                composers = disc.get_composers()
                break
        print(f'Tytul plyty: {title}')
        for i in range(len(songs)):
            print(f'utwor: {songs[i]}, kompozytor: {composers[i]}')
            
    def get_song_in_disc_properties(self, disc_title, song_title):
        result = {}
        for disc in self.lista_plyt:
            if disc_title == disc.get_title():
                result = disc.get_song_properties(song_title)
                break
        for key, val in result.items():
            print(key, ':', val)
            
    def save_to_pickle(self, file_name):
        # zapisanie calej bazy do pliku.pickle
        # plyty beda zawierac liste slownikow [ (disc_info) : [{'title' : '', 'time' : ''....},{song2},{song3}] ]
        backup_baza = []
        for disc in self.lista_plyt:
            backup_baza.append(disc.save_to_pickle_disc())
        
        File = open(file_name, 'wb')
        pickle.dump(backup_baza, File)
        File.close()
        return None
        
    def unfreez_from_pickle(self, file_name):
        # odtworzenie calej bazy danych z pliku.pickle
        File = open(file_name, 'rb')
        backup_baza = pickle.load(File)
        File.close
        for disc in backup_baza:
# disc = {(disc_info) : [{'title' : 'name', 'time' : 60, 'band' : 'name', 'composer' : 'name', 'number' : 1}, {song2}]}
            for disc_properties, song_list in disc.items():
                # disc_properties = ('cool_vibes', 'DVD', '01')
                temp_disc = Plyta(disc_properties[0], disc_properties[1], disc_properties[2])
                for song in song_list:
                    # add_song(title, time, band, composer, number)
                    temp_disc.add_song(song['title'], song['time'], song['band'], song['composer'], song['number'])
                self.add_disc(temp_disc)
        return None
    
    def __str__(self):
        return f'To jest baza plyt, obecnie posaida {len(self.lista_plyt)} plyt'
              
        
        
# ------------------------------------------------------------------------------------
# add_song(title, time, band, composer, number)
cool_vibes = Plyta('cool_vibes', 'DVD', '01')
cool_vibes.add_song('My Calling', 288, 'The Cooties', 'Ethan Edenburg', 5)
cool_vibes.add_song('The Cult of Dionysus', 187, 'The Orion Experience', 'Orion Simprini', 3)
cool_vibes.add_song('Unsteady', 193, 'X Ambassadors', 'Adam Levin', 2)
cool_vibes.add_song('Unstoppable', 192, 'The Score', 'Eddie Anthony', 4)
cool_vibes.add_song('Knee Socks', 257, 'Arctic Monkeys', 'Alex Turner', 1)
#print(cool_vibes.get_songs())

chill = Plyta('chill', 'CD', '02')
chill.add_song('Dreams', 332, 'Nuages', 'Django Reinhardt', 1)
chill.add_song('Train', 185, 'Brick + Mortar', 'John Tacon', 2)
chill.add_song('Here with me', 196, 'Elina', 'Elina Siirala', 3)
chill.add_song('Alone With You', 186, 'Canyon City', 'Paul Johnson', 4)
chill.add_song('Wells', 223, 'Joshua Hyslop', 'Joshua Hyslop', 5)

lofi = Plyta('lofi', 'CD', '03')
lofi.add_song('[oops]', 142, 'potsu', 'Matthew Floyd', 1)
lofi.add_song('The best moment', 82, "j'san", "j'san", 2)
lofi.add_song('Skipping Stones', 252, 'Dan Reynolds, Hans Zimmer', 'Dan Reynolds, Hans Zimmer', 3)
lofi.add_song('Vinyl Boy', 226, 'BROCKBEATS', 'BROCKBEATS', 4)
lofi.add_song('Save Me', 84, 'H E R B', 'Herbert Schiller', 5)


# ------------------------------------------------------------------------------------
Baza1 = Baza_plyt()
Baza1.add_disc(cool_vibes)
Baza1.add_disc(chill)
Baza1.add_disc(lofi)

# ------------------------------------------------------------------------------------
print(Baza1.get_disc_titles())
print()
Baza1.get_properties('lofi')
print()
Baza1.get_properties('chill')
print()
Baza1.get_composers_songs('cool_vibes')
print()
Baza1.get_song_in_disc_properties('lofi', '[oops]')
print()

print()
Baza1.save_to_pickle('Baza2')

Baza2 = Baza_plyt()
print(Baza2)

Baza2.unfreez_from_pickle('Baza2')
print(Baza2)
print(Baza2.get_disc_titles())
```

Zaimplementuj klasę: automatyczna skrzynia biegów. Klasa ma tylko dwa pola prędkość i bieg. Funkcjonalność:

- zwiększ prędkość
- zmninejsz prędkość
- wyświetl prędkość
- wyświetl numer biegu
- bieg wsteczny
    
Zakładamy

- że zmiana biegu następuje wraz ze zmianą prędkości (inne czynniki pomijamy).  
- zmiana na bieg wsteczny lub z biegu wstecznego jest możliwa po zatrzymaniu pojadu i zmianie ręcznej (do tyłu - do przodu).
    
    

```python
class AutomaticGearbox():
    '''
    Objekt AutomaticGearbox ma przedstawiać automatyczną skrzynię biegów
    
    Args:
        predkosc (num)
        bieg (num or 'R')
    Atributes:
        predkosc - przechowuje predkosc
        bieg - przechowuje bieg
    funkcje:
        przyspiesz - przyspiesza o 10 i zwieksza bieg o 1 jeśli to mozliwe
        zwolnij - zwalnia o 10 jesli to mozliwe i redukuje bieg jesli to mozliwe
        wstecz - jesli predkosc wynosi 0 i bieg = 0 to zmienia bieg na 'R' (wstecz)
        do_przodu - jesli bieg = 'R' i predkosc = 0 ro zmienia kierunek jazdy na do przodu (bieg na 0)
        get_predkosc - zwraca aktualna predkosc
        get_bieg - zwraca aktualny bieg
        '''
    
    def __init__(self):
        self.set_predkosc(0)
        self.set_bieg(0)
        
    def set_predkosc(self, x):
        if isinstance(x, int):
            self.predkosc = x
        
    def get_predkosc(self):
        return self.predkosc
    
    def set_bieg(self ,x):
        if isinstance(x, (int, str)):
            self.bieg = x
    
    def get_bieg(self):
        return self.bieg
    
    def przyspiesz(self):
        self.set_predkosc(self.predkosc + 10)
        if self.get_bieg() != 'R':
            self.set_bieg(self.get_bieg() + 1)
    
    def zwolnij(self):
        if self.get_predkosc() == 0:
            print('Nie mozesz juz zmniejszyc predkosci')
        else: 
            self.set_predkosc(self.get_predkosc() - 10)
            
        if self.get_bieg() != 'R' and self.get_bieg() != 0:
            self.set_bieg(self.get_bieg() - 1 )
            
    def wstecz(self):
        if self.get_bieg() == 0:
            if self.get_predkosc() == 0:
                self.set_bieg('R')
        elif self.get_bieg() == 'R':
            print('Pojazd przemieszcza sie do tylu')
        else: 
            print('Zbyt duza predkosc by zmienic bieg na wsteczny')
        
    def do_przodu(self):
        if self.get_bieg() == 'R':
            if self.get_predkosc() == 0:
                self.set_bieg(0)
            else:
                print('Zbyt duza predkosc by zmienic bieg na luz')
        else:
            print('Pojazd przemieszcza sie do przodu')
    
auto1 = AutomaticGearbox()

auto1.przyspiesz()          # predkosc += 10; bieg += 1
auto1.przyspiesz()          # predkosc += 10; bieg += 1
print(auto1.get_predkosc()) # predkosc = 20
print(auto1.get_bieg())     # bieg = 2
auto1.zwolnij()             # predkosc -= 10; bieg -= 1
auto1.wstecz()             # predkosc = 10; bieg = 1, nie może: 'Zbyt duza predkosc by zmienic bieg na wsteczny'
auto1.zwolnij()             # predkosc -= 10; bieg -= 1
auto1.zwolnij()             # predkosc = 0; bieg = 0, nie może: 'Nie mozesz juz zmniejszyc predkosci'
auto1.wstecz()              # predkosc = 0; bieg = 0, może
auto1.przyspiesz()          # predkosc += 10; bieg = 'R'
auto1.do_przodu()           # predkosc = 10; bieg = 'R', nie może: 'Zbyt duza predkosc by zmienic bieg na luz'
auto1.wstecz()              # predkosc = 10; bieg = 'R', nie może: 'Pojazd przemieszcza sie do tylu'
auto1.zwolnij()             # predkosc -= 10; bieg = 'R'
auto1.zwolnij()             # predkosc = 0; bieg = 'R', nie może: 'Nie mozesz juz zmniejszyc predkosci'
auto1.do_przodu()           # predkosc = 0; bieg = 'R', może
auto1.przyspiesz()          # predkosc += 10; bieg += 1
auto1.do_przodu()           # predkosc = 10; bieg = 1, nie może: 'Pojazd przemieszcza sie do przodu'
print(auto1.get_predkosc()) # predkosc = 10
print(auto1.get_bieg())     # bieg = 1
```
